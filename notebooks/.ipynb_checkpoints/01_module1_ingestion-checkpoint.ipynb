## Cell 4 — Sentinel-1 SAR functions

Defines the SAR processing pipeline. **Nothing runs yet** — these are function definitions.

**Physics:** Oil on water dampens capillary waves, reducing radar backscatter. Oil slicks appear as anomalously **dark patches** in VV polarisation images. The Lee speckle filter reduces salt-and-pepper noise before dark spot detection.

```python
# ── Sentinel-1 SAR ingestion and processing functions ────────────────────

def get_s1_collection(roi, start, end):
    """
    Query Sentinel-1 IW GRD scenes.
    IW mode = 250km swath, 10m resolution.
    VV polarisation = best contrast for oil-on-water detection.
    DESCENDING orbit = more consistent geometry over Niger Delta.
    """
    return (
        ee.ImageCollection('COPERNICUS/S1_GRD')
        .filterBounds(roi)
        .filterDate(start, end)
        .filter(ee.Filter.eq('instrumentMode', 'IW'))
        .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV'))
        .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VH'))
        .filter(ee.Filter.eq('orbitProperties_pass', 'DESCENDING'))
        .select(['VV', 'VH'])
    )


def apply_lee_speckle_filter(image, kernel_size=7):
    """
    Lee adaptive speckle filter.
    Reduces multiplicative SAR noise while preserving edges.
    kernel_size=7 is standard for oil spill detection (ESA guidelines).
    """
    def filter_band(band_name):
        band   = image.select([band_name])
        kernel = ee.Kernel.square(kernel_size // 2)
        b_mean = band.reduceNeighborhood(reducer=ee.Reducer.mean(), kernel=kernel)
        b_var  = band.reduceNeighborhood(reducer=ee.Reducer.variance(), kernel=kernel)
        noise  = ee.Number(
            b_var.reduceRegion(
                reducer=ee.Reducer.mean(), geometry=image.geometry(),
                scale=500, maxPixels=1e8, bestEffort=True
            ).values().get(0)
        )
        weight   = b_var.divide(b_var.add(noise))
        filtered = b_mean.add(weight.multiply(band.subtract(b_mean)))
        return filtered.rename(band_name)

    return (
        filter_band('VV')
        .addBands(filter_band('VH'))
        .copyProperties(image, ['system:time_start', 'system:index'])
    )


def compute_sar_features(image, roi, sigma_threshold=1.5):
    """
    Add feature bands for oil spill detection:
      dark_spot_mask      : 1 where VV is anomalously low
      dark_spot_magnitude : how many sigma below mean
      vv_vh_ratio         : discriminates oil from wind shadow
      cross_pol_diff      : VV minus VH (dB)
    """
    vv = image.select('VV')
    vh = image.select('VH')

    stats = vv.reduceRegion(
        reducer=ee.Reducer.mean().combine(ee.Reducer.stdDev(), sharedInputs=True),
        geometry=roi, scale=100, maxPixels=1e9, bestEffort=True
    )
    mean_vv = ee.Number(stats.get('VV_mean'))
    std_vv  = ee.Number(stats.get('VV_stdDev'))

    threshold  = mean_vv.subtract(std_vv.multiply(sigma_threshold))
    dark_mask  = vv.lt(threshold).rename('dark_spot_mask')
    dark_mag   = mean_vv.subtract(vv).divide(std_vv).rename('dark_spot_magnitude')
    vv_lin     = ee.Image(10).pow(vv.divide(10))
    vh_lin     = ee.Image(10).pow(vh.divide(10))
    ratio      = vv_lin.divide(vh_lin).rename('vv_vh_ratio')
    cross_diff = vv.subtract(vh).rename('cross_pol_diff')

    return (
        image.addBands(dark_mask).addBands(dark_mag)
             .addBands(ratio).addBands(cross_diff)
    )


def build_s1_composite(roi, start, end, apply_filter=True, sigma_threshold=1.5):
    """
    Build a temporal median composite.
    Median suppresses transient ships, rain, and wind roughness —
    leaving persistent dark features (chronic spill zones) visible.
    """
    col   = get_s1_collection(roi, start, end)
    count = col.size().getInfo()
    print(f'  S1 scenes ({start} to {end}): {count}')
    if count == 0:
        print('  WARNING: no scenes found for this ROI/date range')
        return None, None, 0
    if apply_filter:
        col = col.map(apply_lee_speckle_filter)
    col = col.map(lambda img: compute_sar_features(img, roi, sigma_threshold))
    return col.median().clip(roi), col, count


def extract_dark_spot_vectors(composite, roi, min_area_m2=10000):
    """Convert dark spot raster mask to polygon vectors with attributes."""
    vectors = (
        composite.select('dark_spot_mask').selfMask()
        .reduceToVectors(
            geometry=roi, scale=40, geometryType='polygon',
            eightConnected=True, maxPixels=1e9, bestEffort=True
        )
    )
    def annotate(f):
        geom = f.geometry()
        vv_stats  = composite.select('VV').reduceRegion(
            reducer=ee.Reducer.mean(), geometry=geom, scale=40, maxPixels=1e6)
        mag_stats = composite.select('dark_spot_magnitude').reduceRegion(
            reducer=ee.Reducer.mean(), geometry=geom, scale=40, maxPixels=1e6)
        return f.set({
            'mean_vv_db':          vv_stats.get('VV'),
            'dark_spot_magnitude': mag_stats.get('dark_spot_magnitude'),
            'area_m2':             geom.area(maxError=10),
            'signal_type':         'SAR_dark_spot',
            'confidence':          'nominal',
        })
    return vectors.map(annotate).filter(ee.Filter.gte('area_m2', min_area_m2))


print('Sentinel-1 functions defined.')
```

## Cell 5 — FIRMS/VIIRS and TROPOMI SO₂ functions

**VIIRS** detects fire radiative power day and night — artisanal refineries burn continuously.

**TROPOMI** detects SO₂ column density — crude oil cracking releases sulfur compounds even through cloud cover (UV backscatter). Persistent SO₂ > 3 DU over the same location is a strong illegal refinery signal.

**Nothing runs yet** — these are function definitions.

```python
# ── FIRMS / VIIRS fire hotspot functions ─────────────────────────────────

def get_firms_collection(roi, start, end):
    """FIRMS VIIRS 375m active fire detections from GEE."""
    return (
        ee.ImageCollection('FIRMS')
        .filterDate(start, end)
        .filterBounds(roi)
        .select(['T21', 'confidence', 'line_number'])
    )


def compute_firms_composite(col, roi):
    """
    Build fire persistence composite.
    T21_max   : peak brightness temp — captures hottest events
    fire_count: detection frequency — persistent sources score high
    T21_mean  : average — stable mean = persistent low-intensity source
    """
    return (
        col.select('T21').max().rename('T21_max').clip(roi)
        .addBands(col.select('T21').count().rename('fire_count').clip(roi))
        .addBands(col.select('T21').mean().rename('T21_mean').clip(roi))
    )


def extract_fire_hotspots(firms_comp, roi, t21_threshold_k=330.0, min_count=3):
    """
    Extract persistent hotspots as point features.
    Filters: T21 > 330K AND detected on >= 3 separate dates.
    Classifies as: refinery_candidate / flare_candidate / agricultural_or_other
    """
    hot_mask = (
        firms_comp.select('T21_max').gt(t21_threshold_k)
        .And(firms_comp.select('fire_count').gte(min_count))
    )
    vectors = (
        firms_comp.select('T21_max').updateMask(hot_mask)
        .reduceToVectors(
            geometry=roi, scale=375, geometryType='centroid',
            eightConnected=True, maxPixels=1e9, bestEffort=True
        )
    )
    def annotate(f):
        geom  = f.geometry()
        stats = firms_comp.reduceRegion(
            reducer=ee.Reducer.mean().combine(ee.Reducer.max(), sharedInputs=True),
            geometry=geom.buffer(500), scale=375, maxPixels=1e6
        )
        t21_max  = ee.Number(stats.get('T21_max_max'))
        count    = ee.Number(stats.get('fire_count_mean'))
        source   = ee.Algorithms.If(
            t21_max.gt(380), 'flare_candidate',
            ee.Algorithms.If(count.gte(10), 'refinery_candidate', 'agricultural_or_other')
        )
        return f.set({
            'T21_max_K':     t21_max,
            'T21_mean_K':    ee.Number(stats.get('T21_mean_mean')),
            'fire_count':    count,
            'signal_type':   'FIRMS_fire_hotspot',
            'likely_source': source,
            'confidence':    ee.Algorithms.If(count.gte(10), 'high', 'nominal'),
        })
    return vectors.map(annotate)


# ── TROPOMI SO₂ functions ────────────────────────────────────────────────

def get_tropomi_collection(roi, start, end):
    """
    Sentinel-5P TROPOMI SO₂ total column density.
    Resolution: 5.5 x 3.5 km. Daily revisit.
    Band: SO2_column_number_density in mol/m²
    Convert to Dobson Units: mol/m² x 2241.5
    Niger Delta background: ~0.5-1.5 DU
    Illegal refinery signal: 3-50+ DU
    """
    return (
        ee.ImageCollection('COPERNICUS/S5P/NRTI/L3_SO2')
        .filterDate(start, end)
        .filterBounds(roi)
        .select(['SO2_column_number_density', 'cloud_fraction'])
    )


def compute_so2_composite(col, roi, max_cloud=0.3):
    """
    Cloud-masked SO₂ composite in Dobson Units.
    cloud_fraction > 0.3 → retrieval unreliable → masked out.
    """
    def mask_and_convert(image):
        cloud  = image.select('cloud_fraction')
        so2_du = (
            image.select('SO2_column_number_density')
            .multiply(2241.5)
            .rename('SO2_DU')
        )
        return so2_du.updateMask(cloud.lt(max_cloud))

    clean = col.map(mask_and_convert)
    return (
        clean.mean().rename('SO2_mean_DU').clip(roi)
        .addBands(clean.max().rename('SO2_max_DU').clip(roi))
        .addBands(clean.reduce(ee.Reducer.percentile([75])).rename('SO2_p75_DU').clip(roi))
        .addBands(clean.count().rename('SO2_obs_count').clip(roi))
    )


def extract_so2_anomalies(so2_comp, roi, threshold_du=3.0, min_obs=5):
    """Extract SO₂ anomaly zones as polygon features."""
    mask    = (
        so2_comp.select('SO2_mean_DU').gt(threshold_du)
        .And(so2_comp.select('SO2_obs_count').gte(min_obs))
    )
    vectors = (
        so2_comp.select('SO2_mean_DU').updateMask(mask)
        .reduceToVectors(
            geometry=roi, scale=5500, geometryType='polygon',
            eightConnected=True, maxPixels=1e9, bestEffort=True
        )
    )
    def annotate(f):
        geom  = f.geometry()
        stats = so2_comp.reduceRegion(
            reducer=ee.Reducer.mean().combine(ee.Reducer.max(), sharedInputs=True),
            geometry=geom, scale=5500, maxPixels=1e6
        )
        so2_mean = ee.Number(stats.get('SO2_mean_DU_mean'))
        conf = ee.Algorithms.If(
            so2_mean.gt(8), 'high',
            ee.Algorithms.If(so2_mean.gt(5), 'nominal', 'low')
        )
        return f.set({
            'SO2_mean_DU': so2_mean,
            'SO2_max_DU':  ee.Number(stats.get('SO2_max_DU_max')),
            'obs_count':   ee.Number(stats.get('SO2_obs_count_mean')),
            'area_m2':     geom.area(maxError=100),
            'signal_type': 'TROPOMI_SO2_anomaly',
            'confidence':  conf,
        })
    return vectors.map(annotate)


def compute_fire_gas_risk_score(so2_comp, firms_comp, roi,
                                 so2_threshold=3.0, t21_threshold=330.0):
    """
    Combined 0-3 illegal refinery risk score.
    0 = no signal
    1 = fire only  (could be agricultural/flare)
    2 = SO2 only   (could be distant industrial)
    3 = fire + SO2 co-located → high-confidence illegal refinery
    """
    so2_mask  = so2_comp.select('SO2_mean_DU').gt(so2_threshold)
    fire_mask = firms_comp.select('T21_max').gt(t21_threshold)
    fire_rs   = fire_mask.reproject(crs='EPSG:4326', scale=5500)
    so2_rs    = so2_mask.reproject(crs='EPSG:4326', scale=5500)
    return (
        so2_rs.multiply(2).add(fire_rs.multiply(1))
        .rename('fire_gas_risk_score').clip(roi)
    )


print('FIRMS/VIIRS and TROPOMI functions defined.')
```

## Cell 6 — Run Sentinel-1 SAR ingestion
Runs the SAR pipeline on GEE servers. Expect ~20 seconds.

```python
print('=== Sentinel-1 SAR ===')

print(f'Baseline ({BASELINE_START} to {BASELINE_END})...')
s1_baseline, s1_baseline_col, n_baseline = build_s1_composite(
    ROI, BASELINE_START, BASELINE_END,
    apply_filter=True, sigma_threshold=SAR_DARK_SPOT_SIGMA
)

print(f'Recent ({RECENT_START} to {RECENT_END})...')
s1_recent, s1_recent_col, n_recent = build_s1_composite(
    ROI, RECENT_START, RECENT_END,
    apply_filter=True, sigma_threshold=SAR_DARK_SPOT_SIGMA
)

print('Extracting dark spot candidate polygons...')
dark_spots = extract_dark_spot_vectors(s1_recent, ROI, min_area_m2=10000)
n_spots    = dark_spots.size().getInfo()

print()
print(f'Baseline scenes : {n_baseline}')
print(f'Recent scenes   : {n_recent}')
print(f'Dark spot zones : {n_spots}')
```

## Cell 7 — Run FIRMS/VIIRS fire hotspot ingestion

```python
print('=== FIRMS / VIIRS ===')

print(f'Baseline ({BASELINE_START} to {BASELINE_END})...')
firms_baseline_col  = get_firms_collection(ROI, BASELINE_START, BASELINE_END)
n_firms_base        = firms_baseline_col.size().getInfo()
firms_baseline_comp = compute_firms_composite(firms_baseline_col, ROI)
print(f'  Images: {n_firms_base}')

print(f'Recent ({RECENT_START} to {RECENT_END})...')
firms_recent_col    = get_firms_collection(ROI, RECENT_START, RECENT_END)
n_firms_rec         = firms_recent_col.size().getInfo()
firms_recent_comp   = compute_firms_composite(firms_recent_col, ROI)
print(f'  Images: {n_firms_rec}')

print('Extracting persistent hotspots...')
fire_hotspots = extract_fire_hotspots(
    firms_recent_comp, ROI,
    t21_threshold_k=FIRMS_BRIGHTNESS_K, min_count=3
)
n_hotspots = fire_hotspots.size().getInfo()
print(f'Persistent hotspots: {n_hotspots}')

if n_hotspots > 0:
    sample = fire_hotspots.limit(3).getInfo()
    print()
    for feat in sample['features']:
        p = feat['properties']
        print(f"  T21_max={p.get('T21_max_K', '?'):.0f}K  "
              f"count={p.get('fire_count', '?'):.0f}  "
              f"source={p.get('likely_source', '?')}")
```

## Cell 8 — Run TROPOMI SO₂ ingestion
The chemical fingerprint of illegal crude burning — works through clouds.

```python
print('=== TROPOMI SO2 ===')

print(f'Baseline ({BASELINE_START} to {BASELINE_END})...')
tropomi_base_col = get_tropomi_collection(ROI, BASELINE_START, BASELINE_END)
n_tropomi_base   = tropomi_base_col.size().getInfo()
so2_baseline     = compute_so2_composite(tropomi_base_col, ROI)
print(f'  Images: {n_tropomi_base}')

print(f'Recent ({RECENT_START} to {RECENT_END})...')
tropomi_rec_col  = get_tropomi_collection(ROI, RECENT_START, RECENT_END)
n_tropomi_rec    = tropomi_rec_col.size().getInfo()
so2_recent       = compute_so2_composite(tropomi_rec_col, ROI)
print(f'  Images: {n_tropomi_rec}')

print(f'Extracting SO2 anomaly zones (> {SO2_THRESHOLD_DU} DU)...')
so2_anomalies = extract_so2_anomalies(
    so2_recent, ROI, threshold_du=SO2_THRESHOLD_DU, min_obs=5
)
n_so2 = so2_anomalies.size().getInfo()
print(f'SO2 anomaly zones: {n_so2}')

print('Computing combined fire + SO2 risk score...')
fire_gas_score = compute_fire_gas_risk_score(
    so2_recent, firms_recent_comp, ROI,
    so2_threshold=SO2_THRESHOLD_DU, t21_threshold=FIRMS_BRIGHTNESS_K
)
print('Score ready  (0=none  1=fire only  2=SO2 only  3=fire+SO2)')
```

## Cell 10 — Interactive risk map

All three sensor layers on a satellite basemap. Use the **layer control (top-right)** to toggle signals on and off.

> Zoom into creek networks — look for SAR dark spots near waterways and co-located SO₂ elevation over fixed locations.

```python
import geemap.foliumap as geemap

m = geemap.Map(center=ROI_CENTER, zoom=ROI_ZOOM)
m.add_basemap('SATELLITE')

# SAR VV backscatter — grayscale
m.addLayer(
    s1_recent.select('VV'),
    {'min': -25, 'max': 0,
     'palette': ['000000','404040','808080','c0c0c0','ffffff']},
    'S1 VV backscatter (dB)'
)

# SAR dark spots — red overlay
m.addLayer(
    s1_recent.select('dark_spot_mask').selfMask(),
    {'palette': ['E24B4A'], 'opacity': 0.7},
    'SAR dark spots — oil spill candidates'
)

# VIIRS max brightness temp
m.addLayer(
    firms_recent_comp.select('T21_max'),
    {'min': 300, 'max': 420,
     'palette': ['E6F1FB','FAC775','EF9F27','E24B4A','791F1F']},
    'VIIRS max brightness temp (K)'
)

# TROPOMI SO2
m.addLayer(
    so2_recent.select('SO2_mean_DU'),
    {'min': 0, 'max': 15,
     'palette': ['E1F5EE','9FE1CB','1D9E75','085041','04342C']},
    'TROPOMI SO2 mean (Dobson Units)'
)

# Combined risk score
m.addLayer(
    fire_gas_score.selfMask(),
    {'min': 1, 'max': 3,
     'palette': ['FAC775','EF9F27','E24B4A'], 'opacity': 0.75},
    'Combined risk score (1=fire  2=SO2  3=both)'
)

# Baseline dark spots for comparison
m.addLayer(
    s1_baseline.select('dark_spot_mask').selfMask(),
    {'palette': ['378ADD'], 'opacity': 0.5},
    'SAR dark spots — BASELINE (comparison)'
)

# ROI boundary
m.addLayer(
    ee.FeatureCollection([ee.Feature(ROI)]),
    {'color': '185FA5', 'fillColor': '00000000'},
    'TNP corridor boundary'
)

m.add_layer_control()
m.add_legend(
    title='PipelineWatch-NG Signals',
    legend_dict={
        'SAR dark spot (oil)':    'E24B4A',
        'VIIRS fire hotspot':     'EF9F27',
        'TROPOMI SO2 anomaly':    '1D9E75',
        'High risk (fire + SO2)': 'A32D2D',
        'Baseline dark spots':    '378ADD',
    },
    position='bottomright'
)

print('Map ready — layer control is top-right corner.')
m
```

## Cell 11 — Fire hotspot time series
Shows whether illegal refining activity is increasing, stable, or seasonal. Expect ~45 seconds.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
import os

def get_daily_fire_counts(col, roi, threshold_k=330.0):
    """Count hot pixels per day across the ROI."""
    def count_day(image):
        date  = image.date().format('YYYY-MM-dd')
        count = (
            image.select('T21').gt(threshold_k)
            .reduceRegion(reducer=ee.Reducer.sum(), geometry=roi,
                          scale=375, maxPixels=1e9, bestEffort=True)
            .get('T21')
        )
        return ee.Feature(None, {'date': date, 'fire_pixels': count})

    records = col.map(count_day).getInfo()['features']
    df = pd.DataFrame([
        {'date': r['properties']['date'],
         'fire_pixels': r['properties'].get('fire_pixels', 0) or 0}
        for r in records
    ])
    df['date'] = pd.to_datetime(df['date'])
    return df.set_index('date').sort_index()

print('Extracting fire counts (this takes ~45s)...')
df_base = get_daily_fire_counts(firms_baseline_col, ROI, FIRMS_BRIGHTNESS_K)
df_rec  = get_daily_fire_counts(firms_recent_col,   ROI, FIRMS_BRIGHTNESS_K)
print('Done.')

fig, axes = plt.subplots(2, 1, figsize=(13, 7), sharex=False)
fig.suptitle(
    'PipelineWatch-NG — VIIRS Fire Hotspot Time Series\n'
    'Trans Niger Pipeline corridor, Niger Delta',
    fontsize=13
)

for ax, df, color, label in [
    (axes[0], df_base, '#378ADD',
     f'Baseline  ({BASELINE_START}  to  {BASELINE_END})'),
    (axes[1], df_rec,  '#E24B4A',
     f'Recent    ({RECENT_START}  to  {RECENT_END})'),
]:
    rolled = df['fire_pixels'].rolling(7, min_periods=1).mean()
    ax.bar(df.index, df['fire_pixels'], color=color, alpha=0.25, width=1)
    ax.plot(df.index, rolled, color=color, linewidth=2, label='7-day rolling mean')
    mean_val = df['fire_pixels'].mean()
    ax.axhline(mean_val, color=color, linewidth=1, linestyle='--', alpha=0.6)
    ax.text(df.index[0], mean_val * 1.08, f'mean = {mean_val:.1f}',
            color=color, fontsize=9)
    ax.set_ylabel('Hot pixels (T21 > 330 K)')
    ax.set_title(label, fontsize=11)
    ax.legend(loc='upper right', fontsize=9)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
    ax.xaxis.set_major_locator(mdates.MonthLocator())
    ax.grid(axis='y', alpha=0.3)
    ax.tick_params(axis='x', rotation=30)

plt.tight_layout()
os.makedirs('../outputs', exist_ok=True)
plt.savefig('../outputs/m1_fire_timeseries.png', dpi=150, bbox_inches='tight')
print('Saved to outputs/m1_fire_timeseries.png')
plt.show()
```

## Cell 12 — SO₂ baseline vs recent comparison

```python
import numpy as np

def get_so2_stats(comp, roi):
    return comp.select('SO2_mean_DU').reduceRegion(
        reducer=(
            ee.Reducer.mean()
            .combine(ee.Reducer.max(), sharedInputs=True)
            .combine(ee.Reducer.percentile([50, 75]), sharedInputs=True)
        ),
        geometry=roi, scale=5500, maxPixels=1e9, bestEffort=True
    ).getInfo()

print('Computing SO2 statistics...')
base_s = get_so2_stats(so2_baseline, ROI)
rec_s  = get_so2_stats(so2_recent,   ROI)

metrics   = ['Mean', 'Median (p50)', 'p75', 'Max']
keys      = ['SO2_mean_DU_mean', 'SO2_mean_DU_p50',
             'SO2_mean_DU_p75',  'SO2_mean_DU_max']
base_vals = [base_s.get(k, 0) or 0 for k in keys]
rec_vals  = [rec_s.get(k, 0)  or 0 for k in keys]

x, w = np.arange(len(metrics)), 0.35
fig, ax = plt.subplots(figsize=(9, 5))
fig.suptitle(
    'PipelineWatch-NG — TROPOMI SO₂ Baseline vs Recent\n'
    'Trans Niger Pipeline corridor',
    fontsize=12
)

ax.bar(x - w/2, base_vals, w,
       label=f'Baseline  ({BASELINE_START[:7]} to {BASELINE_END[:7]})',
       color='#378ADD', alpha=0.85)
ax.bar(x + w/2, rec_vals, w,
       label=f'Recent    ({RECENT_START[:7]} to {RECENT_END[:7]})',
       color='#E24B4A', alpha=0.85)
ax.axhline(SO2_THRESHOLD_DU, color='#854F0B', linewidth=1.5,
           linestyle='--', label=f'Anomaly threshold ({SO2_THRESHOLD_DU} DU)')

ax.set_ylabel('SO2 column density (Dobson Units)')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend(fontsize=9)
ax.grid(axis='y', alpha=0.3)

for bar in ax.patches:
    h = bar.get_height()
    if h > 0.01:
        ax.text(bar.get_x() + bar.get_width() / 2, h + 0.03,
                f'{h:.2f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.savefig('../outputs/m1_so2_comparison.png', dpi=150, bbox_inches='tight')
print('Saved to outputs/m1_so2_comparison.png')
plt.show()
```

## Cell 13 — Export GeoJSON for dashboard
Writes small GeoJSON files to `data/cached/`. Commit these to GitHub — the Streamlit dashboard reads them directly so it needs no GEE API calls at demo time.

```python
import json
from datetime import datetime

CACHE_DIR = '../data/cached'
os.makedirs(CACHE_DIR, exist_ok=True)

def export_fc(fc, filename, label, max_features=500):
    path = os.path.join(CACHE_DIR, filename)
    n    = fc.size().getInfo()
    if n == 0:
        print(f'  {label}: 0 features — writing empty file')
        with open(path, 'w') as f:
            json.dump({'type': 'FeatureCollection', 'features': []}, f)
        return
    print(f'  {label}: {n} features...')
    gj = fc.limit(max_features).getInfo()
    gj['metadata'] = {
        'exported': datetime.now().isoformat(),
        'count':    n,
        'signal':   label,
        'period':   f'{RECENT_START} to {RECENT_END}',
    }
    with open(path, 'w') as f:
        json.dump(gj, f, indent=2)
    kb = os.path.getsize(path) / 1024
    print(f'    saved: {path}  ({kb:.1f} KB)')

print('=== Exporting GeoJSON ===')
export_fc(dark_spots,    'sar_dark_spots.geojson',  'SAR dark spots')
export_fc(fire_hotspots, 'fire_hotspots.geojson',   'VIIRS fire hotspots')
export_fc(so2_anomalies, 'so2_anomalies.geojson',   'TROPOMI SO2 anomalies')

meta = {
    'module':     'M1_ingestion',
    'exported':   datetime.now().isoformat(),
    'study_area': 'Trans Niger Pipeline corridor, Niger Delta',
    'bbox':       ROI_BOUNDS,
    'detections': {
        'sar_dark_spots': n_spots,
        'fire_hotspots':  n_hotspots,
        'so2_anomalies':  n_so2,
    },
}
with open(os.path.join(CACHE_DIR, 'm1_metadata.json'), 'w') as f:
    json.dump(meta, f, indent=2)
print('  metadata: data/cached/m1_metadata.json')
print()
print('All outputs written. Commit data/cached/ to GitHub for the dashboard.')
```
